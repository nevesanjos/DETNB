```python
from obspy.core import Stream, read
from obspy.signal.trigger import coincidence_trigger
from obspy.signal.trigger import classic_sta_lta

from pprint import pprint
import glob
import os

# Pasta com os possveis eventos recortados 
path = "/home/felipe/tmp/Separando_eventos/"

# trigger_on = 5; trigger_off = 2; sta = 1; lta = 10

trigger_on = 5
trigger_off = 2
sta = 0.1
lta = 20

############################################################
# Dia juliano
j = "116"

# Caminho para os Dados
dados = "/home/felipe/sources/DADOS/seiscomp_NRT/2022/"

# Rede e estaes
rede = ['LS', 'NB']
#estacoes = ['NBLV', 'NBTA', 'NBAN', 'NBRFO']
estacoes = ['JAC', 'NBPN']
############################################################

st = Stream()
files = []
for r in rede:
    files = files + glob.glob(dados + r + "/*/*HHZ*/*" + j)

for f in files:
    for e in estacoes:
        if(e in f):
            tr = read(f)
            df = tr[0].stats.sampling_rate
            if(df > 100):
                tr.decimate(4)
            st += tr

st2 = st.copy()

st2.detrend()
st2.filter('bandpass', freqmin=2, freqmax=20)

trig = coincidence_trigger("recstalta", trigger_on, trigger_off, st2, 2, sta=sta, lta=lta,
                          details=True)
pprint(trig)
```

```python
for d in trig:
    start = d['time'] - 5
    end = start + 15
    
    pasta = path + "eventos/" + j
    if not os.path.exists(pasta):
        os.mkdir(pasta)
    pasta = path + "eventos/" + j + "/" + str(d['time'])
    if not os.path.exists(pasta):
        os.mkdir(pasta)
        
    for s in d['stations']:
        #print(d['time'])
        for f in files:
            if (s in f):
                st = read(f)
                tr = st.copy()
                tr.trim(start, end)
                tr.write(pasta + "/" + str(s) + ".HHZ", format='SAC')
                
                ff = f.replace('HHZ','HHN')
                st = read(ff)
                tr = st.copy()
                tr.trim(start, end)
                tr.write(pasta + "/"+str(s)+".HHN", format='SAC')
                
                ff = f.replace('HHZ','HHE')
                ff.replace('HHN','HHE')
                st = read(ff)
                tr = st.copy()
                tr.trim(start, end)
                tr.write(pasta + "/" + str(s)+".HHE", format='SAC')
                
                
                #tr.plot()
```
